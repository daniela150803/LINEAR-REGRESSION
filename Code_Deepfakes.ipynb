```
import tensorflow as tf
from tensorflow.keras import layers, optimizers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Dense, GlobalAveragePooling2D, Dropout, Conv2D,
    BatchNormalization, MaxPooling2D, Concatenate
)
import numpy as np
import os
import matplotlib.pyplot as plt
import pathlib
from sklearn.metrics import confusion_matrix, classification_report, f1_score
import seaborn as sns
from collections import Counter

##########################################
# 1. Ajusta la ruta principal a tu dataset
##########################################
# /kaggle/input/deepfake-and-real-images/Dataset
# y dentro hay:
#   Train/
#     Fake/
#     Real/
#   Validation/
#     Fake/
#     Real/
#   Test/
#     Fake/
#     Real/
data_dir = "/kaggle/input/deepfake-and-real-images/Dataset"

train_dir = os.path.join(data_dir, "Train")
val_dir   = os.path.join(data_dir, "Validation")
test_dir  = os.path.join(data_dir, "Test")

##########################################
# 2. Parámetros de configuración
##########################################
img_height = 224
img_width = 224
batch_size = 8   # Tamaño de batch bajo para ahorrar memoria
AUTOTUNE = tf.data.AUTOTUNE

# Se entrena en dos fases
initial_lr = 0.001
fine_tune_lr = 1e-5
epochs_phase1 = 3
epochs_phase2 = 2

##########################################
# 3. Carga de cada subset con image_dataset_from_directory
##########################################
# Asegúrate de que en train_dir estén las subcarpetas Fake/ y Real/

train_ds_raw = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    label_mode="binary",
    image_size=(img_height, img_width),
    batch_size=batch_size,
    shuffle=True   # Se barajan las imágenes
)

val_ds_raw = tf.keras.preprocessing.image_dataset_from_directory(
    val_dir,
    label_mode="binary",
    image_size=(img_height, img_width),
    batch_size=batch_size,
    shuffle=True
)

test_ds_raw = tf.keras.preprocessing.image_dataset_from_directory(
    test_dir,
    label_mode="binary",
    image_size=(img_height, img_width),
    batch_size=batch_size,
    shuffle=False  # No es necesario barajar test
)

# Imprime las clases detectadas en Train
print("Clases detectadas en Train:", train_ds_raw.class_names)
# Debe mostrar algo como ['Fake', 'Real']

# Prefetch para mejorar el rendimiento, sin usar .cache() para no saturar la RAM
train_ds = train_ds_raw.prefetch(AUTOTUNE)
val_ds   = val_ds_raw.prefetch(AUTOTUNE)
test_ds  = test_ds_raw.prefetch(AUTOTUNE)

##########################################
# 4. Definición de FrequencyAnalyzer y AttentionModule
##########################################
class FrequencyAnalyzer(layers.Layer):
    def __init__(self, **kwargs):
        super(FrequencyAnalyzer, self).__init__(**kwargs)
    def build(self, input_shape):
        # Pocos filtros (16) para no usar demasiada memoria
        self.conv = Conv2D(16, (3, 3), activation='relu', padding='same')
        self.bn = BatchNormalization()
        super(FrequencyAnalyzer, self).build(input_shape)
    def call(self, inputs):
        # Convertir a escala de grises
        weights = tf.constant([0.299, 0.587, 0.114], dtype=inputs.dtype)
        gray = tf.reduce_sum(inputs * weights, axis=-1)
        # FFT 2D
        fft = tf.signal.fft2d(tf.cast(gray, tf.complex64))
        fft_shifted = tf.signal.fftshift(fft)
        magnitude = tf.math.log(tf.abs(fft_shifted) + 1e-6)
        magnitude = tf.expand_dims(magnitude, axis=-1)
        x = self.conv(magnitude)
        x = self.bn(x)
        return x

class AttentionModule(layers.Layer):
    def __init__(self, channels, ratio=8, **kwargs):
        super(AttentionModule, self).__init__(**kwargs)
        self.channels = channels
        self.global_pool = GlobalAveragePooling2D()
        self.dense1 = Dense(channels // ratio, activation='relu')
        self.dense2 = Dense(channels, activation='sigmoid')
    def call(self, inputs):
        x = self.global_pool(inputs)
        x = self.dense1(x)
        x = self.dense2(x)
        x = tf.reshape(x, [-1, 1, 1, self.channels])
        return inputs * x

##########################################
# 5. Definición del modelo (MobileNetV2 + FrequencyAnalyzer)
##########################################
def create_model(input_shape=(224,224,3), freeze_until=100, use_frequency=True):
    inputs = layers.Input(shape=input_shape)
    # Normalizamos [0,1] en la imagen
    x = layers.Rescaling(1./255)(inputs)

    base_model = MobileNetV2(
        include_top=False,
        weights='imagenet',
        input_tensor=x,
        pooling=None
    )
    # Congelamos las primeras 'freeze_until' capas
    for layer in base_model.layers[:freeze_until]:
        layer.trainable = False
    for layer in base_model.layers[freeze_until:]:
        layer.trainable = True

    x_base = base_model.output
    x_base = GlobalAveragePooling2D()(x_base)
    x_base = Dense(128, activation='relu')(x_base)

    if use_frequency:
        x_freq = FrequencyAnalyzer()(x)
        x_freq = MaxPooling2D((2,2))(x_freq)
        x_freq = GlobalAveragePooling2D()(x_freq)
        x_freq = Dense(32, activation='relu')(x_freq)
        x_concat = Concatenate()([x_base, x_freq])
    else:
        x_concat = x_base

    outputs = Dense(1, activation='sigmoid')(x_concat)
    model = Model(inputs=inputs, outputs=outputs)
    return model

##########################################
# 6. Construcción y compilación del modelo
##########################################
model = create_model(freeze_until=100, use_frequency=True)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=initial_lr),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.summary()

##########################################
# 7. Callbacks y entrenamiento en dos fases
##########################################
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
early_stop = EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True)
lr_reduce = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6)

# ---- Fase 1: Entrenamiento de la "cabeza" ----
history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs_phase1,
    callbacks=[early_stop, lr_reduce],
    verbose=1
)

# Descongelamos todas las capas para el fine-tuning completo
for layer in model.layers:
    layer.trainable = True

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=fine_tune_lr),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ---- Fase 2: Fine-tuning completo ----
history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs_phase2,
    callbacks=[early_stop, lr_reduce],
    verbose=1
)

##########################################
# 8. Visualización de historia de entrenamiento
##########################################
import matplotlib.pyplot as plt

acc_phase1 = history1.history['accuracy']
acc_phase2 = history2.history['accuracy']
val_acc_phase1 = history1.history['val_accuracy']
val_acc_phase2 = history2.history['val_accuracy']

loss_phase1 = history1.history['loss']
loss_phase2 = history2.history['loss']
val_loss_phase1 = history1.history['val_loss']
val_loss_phase2 = history2.history['val_loss']

acc_all = acc_phase1 + acc_phase2
val_acc_all = val_acc_phase1 + val_acc_phase2
loss_all = loss_phase1 + loss_phase2
val_loss_all = val_loss_phase1 + val_loss_phase2

plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(acc_all, 'b-', label='Training')
plt.plot(val_acc_all, 'r-', label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(loss_all, 'b-', label='Training')
plt.plot(val_loss_all, 'r-', label='Validation')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

##########################################
# 9. Test Evaluation and Threshold Tuning
##########################################
y_true = []
y_pred = []
all_probs = []

for images, labels in test_ds:
    preds = model.predict(images)
    all_probs.extend(preds.flatten().tolist())
    y_true.extend(labels.numpy().flatten().tolist())
    y_pred.extend((preds.flatten() >= 0.5).astype(int))

print("\nTest Set Real Distribution (0=Fake, 1=Real):", Counter(y_true))
print("Predictions with threshold=0.5:", Counter(y_pred))

print("\nClassification Report (threshold=0.5):")
print(classification_report(y_true, y_pred, target_names=['Fake','Real']))

# Optimal threshold search based on F1-score
import numpy as np
best_thr = 0.5
best_f1 = 0.0
for t in np.linspace(0, 1, 51):
    y_pred_thr = (np.array(all_probs) >= t).astype(int)
    current_f1 = f1_score(y_true, y_pred_thr)
    if current_f1 > best_f1:
        best_f1 = current_f1
        best_thr = t

print(f"Best Threshold Found: {best_thr:.2f} with F1-score: {best_f1:.4f}")
y_pred_best = (np.array(all_probs) >= best_thr).astype(int)

print(f"\nClassification Report (threshold={best_thr:.2f}):")
print(classification_report(y_true, y_pred_best, target_names=['Fake','Real']))

cm = confusion_matrix(y_true, y_pred_best)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Fake','Real'], yticklabels=['Fake','Real'])
plt.xlabel("Prediction")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Test Set")
plt.tight_layout()
plt.show()

##########################################
# 10. Histogram of Predicted Probabilities
##########################################
plt.figure(figsize=(10,4))
plt.hist(all_probs, bins=30)
plt.title("Distribution of Predicted Probabilities on Test Set")
plt.xlabel("Probability (being 'Real')")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
